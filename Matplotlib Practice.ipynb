## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD

plt.plot(x,y,'r-')
plt.show
```

```python

plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

### Multi Plots

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OBJECT ORIENTED METHOD

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.3,0.2])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

### Subplots

```python
fig,axes = plt.subplots()
```

```python
fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(5,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(5,2),nrows=2,ncols=1)


axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

### Colours Setting, Line-width, Linestyle

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#ff8c00') #RGB Hex code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#ff8c00',linewidth=3) #Line width
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#ff8c00',linewidth=3,alpha=0.5) Line visibility
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#ff8c00',linewidth=3,linestyle=':') #Line Style
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#ff8c00',linewidth=3,linestyle='-.') #Line Style
```
