```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Set Title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y, 'r')
ax1.set_title('LARGER PLOT')

ax2.plot(x,y, 'r')
ax2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,y, 'r')
ax1.set_title('LARGER PLOT')

ax2.plot(x,y, 'r')
ax2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,z)
ax1.set_title('Larger Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y, color='purple',linewidth=3,linestyle='-.')
axes[0].set_title('First Plot')

axes[1].plot(x,z, color='red',linewidth=3,linestyle='-')
axes[1].set_title('Second Plot')

plt.tight_layout()
```
